```python
import plotly.express as px, pandas as pd
```

```python
# Load in the data
# Data taken from: https://archive.ics.uci.edu/ml/datasets/Student+Performance
df = pd.read_csv('D:\\Grad School\\Spring 2020\\605.662 - Data Visualization\\Project 4\\data\\student-mat.csv', delimiter=';')
print(df)
```

```python
new_df = df.corr()
print(new_df.shape)
```

```python
# Code influenced from:
    # https://plotly.com/python/heatmaps/
    # https://stackoverflow.com/questions/28006793/pandas-dataframe-to-list-of-lists
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Heatmap(
                    z=new_df.values.tolist(),
                    x=new_df.columns,
                    y=new_df.columns,
                    colorscale='Cividis'))
fig.show()
```
