```python
import plotly.express as px, pandas as pd
```

```python
# Data taken from: https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide
df = pd.read_csv('D:\\Grad School\\Spring 2020\\605.662 - Data Visualization\\Project 4\\data\\covid.csv')
print(df)
```

```python
# Code influenced from:
    # https://plotly.com/python/pie-charts/
    # Colorscheme selected from: https://plotly.com/python/colorscales/
fig = px.pie(df, values='cases', names='countriesAndTerritories', title='COVID-19 Cases by Country')
fig.show()
```

```python
new_df = df[~(df['cases'] <= 1000)]

fig = px.pie(new_df, values='cases', names='countriesAndTerritories', 
             title='COVID-19 Cases by Country',
            color_discrete_sequence=px.colors.sequential.Viridis)
fig.show()
```
